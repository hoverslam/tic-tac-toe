```python
import pandas as pd

import sys
sys.path.append("../game")
from tic_tac_toe import Training
```

```python
games = 100000
state = (1, 2, 1, 0, 2, 0, 0, 0, 0)

t = Training(0.95, 0.1, [1, 0.1, 0.5])
history = t.start(games, history=state)
history = pd.DataFrame(history, columns=["game", "p1", "p2"])
```

```python
# Check if rows have change
history_p1 = history[["game", "p1"]]
history_p1["match"] = history_p1["p1"].eq(history_p1["p1"].shift())

# Only save rows that have changed
changes_p1 = history_p1.loc[history_p1["match"] == False].drop(columns=["match"])

# Unnest p1 in multiple columns
unnest_p1 = pd.DataFrame(changes_p1["p1"].to_list(), columns=["0", "1", "2", "3", "4", "5", "6", "7", "8"])
unnest_p1 = unnest_p1.join(changes_p1["game"].reset_index()).drop(columns=["index"])
first_column = unnest_p1.pop("game")
unnest_p1.insert(0, "game", first_column)
```
